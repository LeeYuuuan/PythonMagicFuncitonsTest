## measure the time for a block

```python
%%time
a = sum(1 for _ in range(1000))
a + 10
```

## measure the time multiple times

```python
%%timeit -n 10
a = sum(1 for _ in range(1000))
a + 10
```

## prun >> profile

```python
def fib(n):
    if n <= 2:
        return 1
    return fib(n-1) + fib(n-2)
```

```python
%%prun
fib(30)
```

```python
# 这个命令能够展现出各个命令的调用次数和花费时间，
#适合分析并优化程序性能。

```
